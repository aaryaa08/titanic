```python
train["Nationality_FR"]=train["Embarked"]=="C"
```

```python
train["Nationality_EN"]= (train["Embarked"] == "S") | (train["Embarked"] == "C")
```

```python
train["FamilySize"] = (train["SibSp"]+train["Parch"])+1
train[["FamilySize", "SibSp", "Parch"]].head()
```

```python
train.loc[train["Sex"] == "male", "Sex"] = 0
train.loc[train["Sex"] == "female", "Sex"] = 1
```

```python
train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"
```

```python
train.loc[train["Age"].isnull(), "Age"] = train["Age"].mean()
```

```python
test.loc[train["Age"].isnull(), "Age"] = test["Age"].mean()
```

```python
test.loc[test["Fare"].isnull(), "Fare"] = test["Fare"].mean()
```

## Binning

```python
train["Fare_Low"] = train["Fare"] < 10
train["Fare_Medium"] = (train["Fare"] >= 10) & (train["Fare"] < 50 )
train["Fare_High"] = train["Fare"] >= 50

train[["Fare", "Fare_Low", "Fare_Medium", "Fare_High"]].head()
```

```python
train.head()
```

```python
feature_name = ["Pclass", "Sex", "Fare", "Embarked_C", "Embarked_Q", "Embarked_S"]
```

```python
x_train = train[feature_name]
```

```python
x_train.head()
```

```python
label_name = "Survived"
```

```python
y_train = train[label_name]
y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth = 3, random_state = 2000)
model
```

```python
model.fit(x_train, y_train)
```

```python
x_test = test[feature_name]
x_test.head()
```

```python
test.loc[test["Sex"] == "male", "Sex"] = 0
test.loc[test["Sex"] == "female", "Sex"] = 1

test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"
```

```python
prediction = model.predict(x_test)
prediction[:2]
```

```python
prediction[:10]
```

```python
test['Survived'] = prediction
```
